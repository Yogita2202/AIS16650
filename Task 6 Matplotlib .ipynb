```python
#Markers
#You can use the keyword argument marker to emphasize each point with a specified marker:
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([2, 4, 12, 2])

plt.plot(ypoints, marker = 'o')
plt.plot(ypoints, marker = '*')
plt.show()
```

```python
#Matplotlib Labels and Title
x = np.array([85, 92, 91, 89, 100, 123, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
#Add a plot title and labels for the x- and y-axis:
x = np.array([85, 92, 91, 89, 100, 123, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
#Set Font Properties for Title and Labels
#You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.
x = np.array([85, 92, 91, 89, 100, 123, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
#Position the Title
#You can use the loc parameter in title() to position the title.
#Legal values are: 'left', 'right', and 'center'. Default value is 'center'.
x = np.array([85, 92, 91, 89, 100, 123, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

```python
#Matplotlib Adding Grid Lines
x = np.array([130, 145, 135, 175, 148, 127, 136, 134, 133, 124])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports data")
plt.xlabel("Max Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
#Matplotlib Subplot
#Display Multiple Plots
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([1, 2, 3, 4])
y = np.array([5, 8, 11, 16])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#scatter plot
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

# Display the plot
plt.show()
```

```python
# Sample data for the first scatter plot
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]

# Sample data for the second scatter plot
x2 = [1, 2, 3, 4, 5]
y2 = [3, 5, 7, 11, 13]

# Create scatter plot for the first data set
plt.scatter(x1, y1, color='blue', label='Data Set 1')

# Create scatter plot for the second data set
plt.scatter(x2, y2, color='red', label='Data Set 2')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Two Data Sets')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
#Matplotlib Bars
x = np.array(["A", "B", "C", "D"])
y = np.array([5, 8, 11, 3])

plt.bar(x,y)
plt.show()
```

```python
#Histogram
#A histogram is a graph showing frequency distributions.
# Generate some random data
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Display the plot
plt.show()
```

```python
# Data for the pie chart
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add a title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```
